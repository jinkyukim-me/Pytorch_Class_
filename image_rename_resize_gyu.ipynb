# Import Module

```python
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
%matplotlib inline
```

# Resize

```python
trans = transforms.Compose([transforms.Resize((28, 28))])

train_data = dsets.ImageFolder(root='menwomen/original_data', transform=trans)
```

```python
for num, value in enumerate(train_data):
    data, label = value
    print(num, data, label)
    
    plt.imshow(data)
    break
    
    
```

# Make Train data and Rename files

```python
for num, value in enumerate(train_data):
    data, label = value
    print(num, data, label)
    
    if (label == 0):
        data.save('menwomen/train_data/men/menpic%d_%d.jpeg'%(num, label))
    else:
        data.save('menwomen/train_data/women/womenpic%d_%d.jpeg'%(num, label))
```

# Make Test data and Rename files

```python
for num, value in enumerate(train_data):
    data, label = value
    print(num, data, label)
    
    if num % 5 ==0:
        if (label == 0):
            data.save('menwomen/test_data/men/menpic%d_%d.jpeg'%(num, label))
        else:
            data.save('menwomen/test_data/women/womenpic%d_%d.jpeg'%(num, label))
```
