```python
import torchvision.datasets as dsets
import torchvision.transforms as transforms
```

```python
trans = transforms.Compose([transforms.Resize((28, 28))])

train_data = dsets.ImageFolder(root='dogscats/original_data', transform=trans)
```

```python
for num, value in enumerate(train_data):
    data, label = value
    print(num, data, label)
    
    if (label == 0):
        data.save('dogscats/train_data/cats/catpic%d_%d.jpeg'%(num, label))
    else:
        data.save('dogscats/train_data/dogs/dogpic%d_%d.jpeg'%(num, label))
```
